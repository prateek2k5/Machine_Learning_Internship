```python
#AI Based Carrer Guidance Project
#The AI-Based Career Guidance System is a machine learning-powered web application designed to help students and job seekers 
#identify suitable career paths based on their skills, interests, academic background, and personal preferences.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('dataset9000.csv')
data #View the given Data
```

```python
#Check whether data have present any null values in any column if there remove them 

data.isnull().sum()
```

```python
#Now using the label encoding here because the Machine was unable to process the columns like geography and Gender as they contain 
#String Value machine can understand only numerical values.
#Label Encoding converts the given columns values into numerical values in alphabetical order 
#EX-Female=0 and Male=1 if more values then so on alphabetically.

from sklearn.preprocessing import LabelEncoder
label_enc = LabelEncoder()
data['Database Fundamentals'] = label_enc.fit_transform(data['Database Fundamentals'])
data['Computer Architecture'] = label_enc.fit_transform(data['Computer Architecture'])
data['Distributed Computing Systems'] = label_enc.fit_transform(data['Distributed Computing Systems'])
data['Cyber Security'] = label_enc.fit_transform(data['Cyber Security'])
data['Networking'] = label_enc.fit_transform(data['Networking'])
data['Software Development'] = label_enc.fit_transform(data['Software Development'])
data['Programming Skills'] = label_enc.fit_transform(data['Programming Skills'])
data['Project Management'] = label_enc.fit_transform(data['Project Management'])
data['Computer Forensics Fundamentals'] = label_enc.fit_transform(data['Computer Forensics Fundamentals'])
data['Technical Communication'] = label_enc.fit_transform(data['Technical Communication'])
data['AI ML'] = label_enc.fit_transform(data['AI ML'])
data['Software Engineering'] = label_enc.fit_transform(data['Software Engineering'])
data['Business Analysis'] = label_enc.fit_transform(data['Business Analysis'])
data['Communication skills'] = label_enc.fit_transform(data['Communication skills'])
data['Data Science'] = label_enc.fit_transform(data['Data Science'])
data['Troubleshooting skills'] = label_enc.fit_transform(data['Troubleshooting skills'])
data['Graphics Designing'] = label_enc.fit_transform(data['Graphics Designing'])

data.head()
```

```python
#Now training the model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Features and target
x = data.drop('Role', axis=1) #This will take all the columns except the Exited Column
y = data['Role'] #It will take the exited column only

# Split and scale
from sklearn.model_selection import train_test_split #For splitting the given data into training and testing data
from sklearn.preprocessing import StandardScaler #For Scaling the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)


#Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
y_pred_rf = rf_model.predict(x_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```
