```python
#Importing an In-Built Data from Sklearn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing #Inbuilt Library
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error,r2_score #r2_score tell the effeciency of the model (0-1)
```

```python
#Load and Explore the data from inbuilt library
data=fetch_california_housing()
print(data.DESCR) #Description of the Data
df=pd.DataFrame(data.data,columns=data.feature_names) #Columns is the name of the column as we made it from Scratch for Supervised Leaning as it needs labeled data
df['MedHouseVal']=data.target
df.head() #5 Starting Data Samples 
```

```python
#Preprocessing of the given datasets
x=df.drop('MedHouseVal',axis=1) #Removing First Column and taking complete data
y=df['MedHouseVal']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
#Linear Regression
lr=LinearRegression()
lr.fit(x_train_scaled,y_train)
y_pred_lr=lr.predict(x_test_scaled)
print("RMSE :",np.sqrt(mean_squared_error(y_test,y_pred_lr)))
print("R^2 Score : ",r2_score(y_test,y_pred_lr))
```

```python
#Polynomial Regression
poly=PolynomialFeatures(degree=2) #Binomial- ax^2+bx+c
x_train_poly=poly.fit_transform(x_train_scaled)
x_test_poly=poly.transform(x_test_scaled)
pr=LinearRegression()
pr.fit(x_train_poly,y_train)
y_pred_pr=pr.predict(x_test_poly)
print("RMSE :",np.sqrt(mean_squared_error(y_test,y_pred_pr)))
print("R^2 Score : ",r2_score(y_test,y_pred_pr))
```

```python
#Ridge Regression
ridge=Ridge(alpha=1.0)
ridge.fit(x_train_scaled,y_train)
y_pred_ridge=ridge.predict(x_test_scaled)
print("RMSE :",np.sqrt(mean_squared_error(y_test,y_pred_ridge)))
print("R^2 Score : ",r2_score(y_test,y_pred_ridge))
```

```python
#Lasso Regression
lasso=Lasso(alpha=0.1)
lasso.fit(x_train_scaled,y_train)
y_pred_lasso=lasso.predict(x_test_scaled)
print("RMSE :",np.sqrt(mean_squared_error(y_test,y_pred_lasso)))
print("R^2 Score : ",r2_score(y_test,y_pred_lasso))
```

```python
#Analysis of the models that works better with our Data

result=pd.DataFrame({
'Model':['Linear Regression','Polynomial Regression','Ridge Regression','Lasso Regression'],
'RMSE':[
np.sqrt(mean_squared_error(y_test,y_pred_lr)),
np.sqrt(mean_squared_error(y_test,y_pred_pr)),
np.sqrt(mean_squared_error(y_test,y_pred_ridge)),
np.sqrt(mean_squared_error(y_test,y_pred_lasso))
],
'R^2 Score':[
r2_score(y_test,y_pred_lr),
r2_score(y_test,y_pred_pr),
r2_score(y_test,y_pred_ridge),
r2_score(y_test,y_pred_lasso)
]})
result
```

```python
#Choosing the best model by testing it on one sample data

testdf=x_test.tail(1)
testdf
testdf_scaled=scaler.transform(testdf)
testdf_poly=poly.transform(testdf_scaled)
test_pr=pr.predict(testdf_poly)
testdf_lasso=lasso.predict(testdf_scaled)
testdf_lin=lr.predict(testdf_scaled)
testdf_ridge=ridge.predict(testdf_scaled)
print(test_pr)
print(testdf_lasso)
print(testdf_lin)
print(testdf_ridge)
```
