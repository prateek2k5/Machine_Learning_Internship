```python
#Students Data
arr=np.array([[71,69,43,56],[78,62,43,56],[78,67,43,56],[78,67,43,56],[78,67,43,56],[72,67,43,56]])
print(arr)
print()

#Sum of marks of each students
b=np.sum(arr,axis=1)
print(b)
print()

#Sorted Array
top_students=np.sort(b)
print(top_students)

#Top 3 Students
print("Topper :",top_students[-1:-4:-1])

#Students that failed (marks<50)
labels=np.where(arr<50,"Fail","Pass")
print()
print(labels)

#Students with marks in between 240-245
mar=b[(b>240) & (b<=245)]
print()
print(mar)
```

```python
#Product Quaterly sales report where row represent product and column represent sale
product=np.array([[4500,3800,2200,1000],
                  [3500,5800,1200,3000],
                  [2500,1800,5200,4000],
                  [1500,2800,2200,2000]])
print(product)
b=np.sum(product,axis=0)
#Total sale of 2 Product
print(b[1])

#3 Quater Sale Sum of all products
c=np.sum(product,axis=1)
print(c[2])

#Top two performing product
c=np.sort(b)
print(c[-1:-3:-1])

#Product with sale above 13000
row,col=np.where(product>=5000)
print(np.vstack((row,col)))
```

```python
#Matrix Multiplication
a=np.array([[10,5,-6],
            [7,-4,6],
            [9,9,1]])
b=np.array([5,-8,10])
inv=np.linalg.inv(a)
mul=np.matmul(inv,b)
print(mul)
```

```python
#Crammer Rule
a=np.array([[10,5,-6],
            [7,-4,6],
            [9,9,0]])
B=np.array([5,-8,10])
D=np.linalg.det(a)
print("Determinante is: ",D)
A1 = a.copy()
A1[:,0] = B
Dx = np.linalg.det(A1)

A2 = a.copy()
A2[:,1] = B
Dy = np.linalg.det(A2)

A3 = a.copy()
A3[:,2] = B
Dz = np.linalg.det(A3)

x = Dx / D
y = Dy / D
z = Dz / D

print("x =", x)
print("y =", y)
print("z =", z)
```
