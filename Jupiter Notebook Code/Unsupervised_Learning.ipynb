```python
#Unsupervised Learning

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
data=pd.read_csv('Mall_Customers.csv')
# Select the 4th and 5th columns (index 3 and 4) from the dataset
# and convert them into a NumPy array to use as input features.
features=data.iloc[:,[3,4]].values
print(features)
```

```python
# Create an empty list to store the WCSS (Within-Cluster Sum of Squares)
# for different numbers of clusters.
wcss_list = []

# Loop through cluster values from 1 to 10
for i in range(1, 11):

    # Create a K-Means model with 'i' clusters.
    # init='k-means++' selects initial centroids intelligently.
    # random_state=0 ensures the same results every time the code runs.
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)

    # Train the K-Means model on the selected features.
    kmeans.fit(features)

    # Store the WCSS (inertia) value for the current number of clusters.
    wcss_list.append(kmeans.inertia_)

# Plot the Elbow Method graph.
# X-axis: Number of clusters (1 to 10)
# Y-axis: Corresponding WCSS values
plt.plot(range(1, 11), wcss_list)

# Display the graph.
plt.show()
```

```python
# Create a K-Means model with 3 clusters.
# init='k-means++' selects the initial centroids intelligently.
# random_state=0 ensures the same output every time the code runs.
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)

# Train the K-Means model on the feature data
# and assign each data point to its nearest cluster.
# The cluster labels (0, 1, or 2) are stored in 'ypred'.
ypred = kmeans.fit_predict(features)

# Display the cluster assigned to each data point.
print(ypred)
```

```python
# Create a scatter plot of the dataset.
# features[:, 0] selects all values from the first feature (X-axis).
# features[:, 1] selects all values from the second feature (Y-axis).
plt.scatter(features[:, 0], features[:, 1])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('K-Means Clustering')
plt.show()
```

```python
#What is DBSCAN?

#DBSCAN is an unsupervised machine learning clustering algorithm that groups together data points that are closely packed (high density)
#and marks points in low-density regions as outliers (noise).
#Unlike K-Means, DBSCAN does not require you to specify the number of clusters beforehand.

#1. eps (ε)
#eps is the maximum distance between two points for them to be considered neighbors.

#2. min_samples (MinPts)
#Minimum number of neighboring points required to form a dense region.

#Types of Points in DBSCAN :
#Core Point: A point that has at least min_samples neighboring points within the eps radius.
#Border Point: A point that has fewer than min_samples neighbors but lies within the eps radius of a Core Point.
#Noise Point (Outlier): A point that is neither a Core Point nor a Border Point, so it does not belong to any cluster.
```

```python
#What is Agglomerative Clustering?

#Agglomerative Clustering is an unsupervised machine learning algorithm that builds clusters by starting with each data point 
#as its own cluster and repeatedly merging the two closest clusters until the desired number of clusters is reached.
#It is also known as Bottom-Up Hierarchical Clustering.

#ex: A   B   C   D   E
#Start with Individual Clusters
#{A} {B} {C} {D} {E}

#Merge the Closest Clusters
#Suppose A and B are closest:
#{AB} {C} {D} {E}

#Merge Again
#Suppose D and E are closest:
#{AB} {C} {DE}

#Continue Merging
#Merge C with AB:
#{ABC} {DE}

#Final Merge
#{ABCDE}
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

#Load Iris dataset from sklearn
data=load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['species']=data.target

#Show the first few rows
df.head()
```

```python
#Feature scaling
scaler=StandardScaler()
x_scaled=scaler.fit_transform(df.drop('species',axis=1))

#Optional: Show Scaled Data
pd.DataFrame(x_scaled,columns=data.feature_names).head()
```

```python
#K-Means Clustering
kmeans=KMeans(n_clusters=3,random_state=42)
kmeans_labels=kmeans.fit_predict(x_scaled)

#Add labels to the dataframe
df['KMeans_Cluster']=kmeans_labels
#Plot K-Means clustering
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['sepal length (cm)'],y=df['sepal width (cm)'], hue=df['KMeans_Cluster'],palette='Set1')
plt.title('K-Means Clustering Results')
plt.show()
```

```python
#Evaluate with silhouette score or metrics
from sklearn.metrics import silhouette_score
print("K-Means Silhouette Score :",silhouette_score(x_scaled, kmeans_labels))
```

```python
#Hierarchial Clustering
hierarchical=AgglomerativeClustering(n_clusters=3)
hierarchical_labels=hierarchical.fit_predict(x_scaled)

#Add labels to the dataframe
df['Hierarchical_Cluster']=hierarchical_labels

#Plot Hierarchical Clustering
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['sepal length (cm)'],y=df['sepal width (cm)'], hue=df['KMeans_Cluster'],palette='Set1')
plt.title('Hierarchical Clustering Results')
plt.show()
```

```python
#Evaluate with silhouette score or metrics
from sklearn.metrics import silhouette_score
print("K-Means Silhouette Score :",silhouette_score(x_scaled,hierarchical_labels))
```

```python
#DBCSAN Clustering
dbscan=DBSCAN(eps=0.5,min_samples=5)
dbscan_labels=dbscan.fit_predict(x_scaled)

#Add labels to the dataframe
df['DBSCAN_Cluster']=dbscan_labels

#Plot Hierarchical Clustering
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['sepal length (cm)'],y=df['sepal width (cm)'], hue=df['DBSCAN_Cluster'],palette='Set1')
plt.title('DBSCAN Clustering Results')
plt.show()
```

```python
#Evaluate with silhouette score or metrics
from sklearn.metrics import silhouette_score
print("DBSCAN Silhouette Score :",silhouette_score(x_scaled,dbscan_labels))
```
