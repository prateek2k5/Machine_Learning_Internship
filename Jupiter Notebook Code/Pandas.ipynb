```python
#Fill missing values

df2['Age']=df2['Age'].fillna(df2['Age'].median())
df2['Income']=df2['Income'].fillna(df2['Income'].median())
df2['Gender']=df2['Gender'].fillna(df2['Gender'].mode()[0])
df2['City']=df2['City'].fillna(df2['City'].mode()[0])
df2['Date_of_Joining']=df2['Date_of_Joining'].fillna('2022-01-01')
df2.isnull().sum()
```

```python
#Check Duplicate Values
df2.duplicated().sum()
```

```python
#Drop Duplicates
df2.drop_duplicates(inplace=True)
df2.duplicated().sum()
```

```python
df2.dtypes
```

```python
#Converting Data_Types
df2['Date_of_Joining']=pd.to_datetime(df2['Date_of_Joining'])
df2['Gender']=df2['Gender'].astype('category')
df2.dtypes
```
