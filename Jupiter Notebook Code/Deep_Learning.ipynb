```python
#What is Deep Learning?
#Deep Learning is a subset of Machine Learning that uses artificial neural networks with multiple layers to learn complex 
#patterns from data.
```

```python
#What is ANN?

#ANN = Artificial Neural Network
#ANN is a Deep Learning model inspired by the working of biological neural networks in the human brain.

#Structure of ANN
#ANN mainly has 3 types of layers:
#Input Layer → Takes input/features.
#Hidden Layer(s) → Learns patterns and relationships.
#Output Layer → Gives final prediction.

#Perceptron in Machine Learning
#Perceptron is the simplest type of Artificial Neural Network (ANN). It is a single artificial neuron mainly used for binary classification.

#How Perceptron Works
#Input → Weights → Weighted Sum + Bias → Activation Function → Output

#Learning Process
#Take input.
#Calculate weighted sum.
#Apply activation function.
#Compare prediction with actual output.
#Calculate error.
#Update weights and bias.


#What is Epoch?
#One epoch means the neural network has seen the complete training dataset one time.
#Why multiple epochs?
#Because a neural network usually doesn't learn everything in one attempt.
```

```python
#Simple Perceptron
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Input data for AND Gate
# x1  x2
x = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Output of AND Gate
y = [0, 0, 0, 1]

# Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create Perceptron model
model = Perceptron()

# Train model
model.fit(x_train, y_train)

# Prediction
y_pred = model.predict(x_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Test with new input
new_data = [[1, 1]]

result = model.predict(new_data)

print("Prediction for [1, 1]:", result[0])
```

```python
#Simple Neural Network by doing Hard Code

import numpy as np

class NeuralNetwork():

    def __init__(self):
        #Seeding for random number generation
        np.random.seed(1)
        #Converting weights to a 3 by 1 matrix with values from -1 to 1 and mean of 0
        self.synaptic_weights=2*np.random.random((3,1))-1

    def sigmoid(self,x):
        #applying the sigmoid function
        return 1/(1+np.exp(-x))

    def sigmoid_derivative(self,x):
        #computing derivative to the sigmoid function
        return x*(1-x)

    def train(self,training_inputs,training_outputs,training_iterations):
        #training the model to make accurate prediction while adjusting the weights
        for iteration in range(training_iterations):
            #siphon the training data vid neuron
            output=self.think(training_inputs)
            
            #Computing error rate for the backpropogation
            error=training_outputs-output
            
            #performing weight adjustment
            adjustments=np.dot(training_inputs.T,error*self.sigmoid_derivative(output))
            self.synaptic_weights+=adjustments

    def think(self,inputs):
        #passing the inputs vid the neuron to get output
        #converting values to float

        inputs=inputs.astype(float)
        output=self.sigmoid(np.dot(inputs, self.synaptic_weights))
        return output
```

```python
if __name__ =="__main__":
    #initializing the neuron class
    neural_network=NeuralNetwork()

    #training data consisting of 4 example----3 Input values and 1 Output
    training_inputs=np.array([[0,0,1],
                            [1,1,1],
                            [1,0,1],
                            [0,0,1]])

    training_outputs=np.array([[0,1,1,0]]).T

    #training taking place
    neural_network.train(training_inputs,training_outputs,15000)

    print("Ending weights after Training : ")
    print(neural_network.synaptic_weights)

    user_input_one=input("User Input One: ")
    user_input_two=input("User Input Two: ")
    user_input_three=input("User Input Three: ")

    print("Considering New Situation: ",user_input_one,user_input_two,user_input_three)
    print("New Output data : ")
    print(neural_network.think(np.array([user_input_one,user_input_two,user_input_three])))
    print("We Did it !!!")
```
